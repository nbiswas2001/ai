```python
import pandas as pd
from sklearn.datasets import load_iris

#Load the iris dataset
iris = load_iris()

# Create a DataFrame using data and feature names
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Top 5 rows
df.head()
```

```python
# Add target variable as a separate column (optional)
df['species'] = iris.target

# Bottom 10 rows
df.tail(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
filtered_rows = df[df['sepal length (cm)'] > 5.8]  # Filter rows where sepal length > 5.8
filtered_rows.head(10)
```

```python
value_counts = df['sepal width (cm)'].value_counts() # Count occurrences of each unique value
value_counts.head()
```

```python
import seaborn as sns
# Create a scatter plot of Sepal Length vs Sepal Width colored by species
plt = sns.scatterplot(
    x="sepal length (cm)",
    y="sepal width (cm)",
    hue="species",
    palette="deep",
    data=df
)

# Customise the legend
handles, labels  =  plt.get_legend_handles_labels()
plt.legend(handles, ['setosa','virginica','versicolor'], loc='lower right')

# Title etc
plt.set(
    title='Iris Sepal Dimensions by Species',
    xlabel = 'Sepal Length (cm)',
    ylabel = 'Sepal Width (cm)'
)
```

```python
# Create a boxplot of Sepal Length by Species
plt = sns.boxplot(
    x = "species",
    y = "sepal length (cm)",
    showmeans=True,  # Show mean with a diamond marker
    data=df
)

# Rotate x labels to prevent overlapping
plt.tick_params(axis='x', labelrotation=45)
plt.set_xticklabels(['setosa','virginica','versicolor'])

plt.set(
    title = 'Sepal Length Distribution by Iris Species',
    xlabel = 'Species',
    ylabel = 'Sepal Length (cm)'
)
```

```python
# Histogram
plt = sns.histplot(data=df['sepal length (cm)'])

# Customize the plot (optional)
plt.set(
    title = 'Distribution of Data',
    xlabel = 'Value',
    ylabel = 'Count'
)
```
